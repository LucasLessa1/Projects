```python
file_path = 'C:/Codigos_VS/Projects/PowerApps/SP_STEADYSTATE_0.F09'  # Adjust the file path
column_names = ["SECTION", "BUS No", "BUS Type", "SECTION CURRENT Magnitude(A)", "SECTION CURRENT Angle(deg.)",
                "TOWER (SHUNT) CURRENT Magnitude(A)", "TOWER (SHUNT) CURRENT Angle(deg.)",
                "TOWER (SHUNT) POTENTIAL Magnitude(V)", "TOWER (SHUNT) POTENTIAL Angle(deg.)"]

columns_to_float = ["SECTION CURRENT Magnitude(A)", "SECTION CURRENT Angle(deg.)",
                        "TOWER (SHUNT) CURRENT Magnitude(A)", "TOWER (SHUNT) CURRENT Angle(deg.)",
                        "TOWER (SHUNT) POTENTIAL Magnitude(V)", "TOWER (SHUNT) POTENTIAL Angle(deg.)"]

string_finder = "                             TERMINAL NO.   2, Term2       "
df = extract_data_to_dataframe(file_path=file_path, column_names=column_names, string_finder=string_finder)
df_float = convert_columns_to_float(df, columns_to_float)
result_df = filtered_section_df = filter_data_by_section(df_float, 'Terminal')

result_df = remove_rows_with_equals(result_df)
result_df.to_csv('LT_DATA.csv')
```
